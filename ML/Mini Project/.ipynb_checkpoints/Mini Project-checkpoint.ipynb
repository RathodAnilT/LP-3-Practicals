```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_data = pd.read_csv(r'C:\Users\Anil Rathod\Desktop\LP III\ML\Mini Project\Titanic-Dataset.csv')
train_data.head()
```

```python
import pandas as pd

# Assuming you already have train_data loaded
# Example: train_data = pd.read_csv('path/to/titanic.csv')

# Step 1: Conditionally drop columns if they exist
columns_to_drop = ['PassengerId', 'Name', 'Ticket', 'Cabin']
train_data = train_data.drop([col for col in columns_to_drop if col in train_data.columns], axis=1)

# Step 2: Fill missing Age values with the mean of 'Age'
if 'Age' in train_data.columns:
    train_data['Age'] = train_data['Age'].fillna(train_data['Age'].mean())

# Step 3: Drop rows where 'Embarked' has missing values
if 'Embarked' in train_data.columns:
    train_data.dropna(subset=['Embarked'], inplace=True)

# Step 4: Map 'Sex' to numeric values ('male' -> 1, 'female' -> 0)
if 'Sex' in train_data.columns:
    train_data['Sex'] = train_data['Sex'].map({'male': 1, 'female': 0})

# Step 5: One-hot encode the 'Embarked' column if it exists
if 'Embarked' in train_data.columns:
    train_data = pd.get_dummies(train_data, columns=['Embarked'])

# Display the first few rows to confirm the changes
print(train_data.head())
```

```python
gender_counts = train_data['Sex'].value_counts()
gender_labels = ['Male', 'Female']

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_labels, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#ff9999'])
plt.title('Gender Distribution on the Titanic')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=train_data)
plt.title('Survival Rate by Gender')
plt.xticks([0, 1], ['Female', 'Male'])
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=train_data)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(train_data['Age'], bins=30, kde=False)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked_C', data=train_data, palette='muted')
plt.title('Count of Passengers Embarked from Each Port (Cherbourg)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked_S', data=train_data, palette='muted')
plt.title('Count of Passengers Embarked from Each Port (Southampton)')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked_Q', data=train_data, palette='muted')
plt.title('Count of Passengers Embarked from Each Port (Queenstown)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Age', data=train_data)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr_matrix = train_data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```
